```python
#initialize lists to hold info
names = []
workplaces = []
cleaned_workplaces = []
```

```python
#BeautifulSoup Scraper

def scrape():
    soup = bs(driver.page_source, "lxml")
    name_results = soup.find_all("span","entity-result__title-text t-16")
    for n in name_results:
        name = n.find("span", attrs={"aria-hidden":True})
        try:
            names.append(name.text)
        except:
            names.append(None)

    workplace_results = soup.find_all("div", "entity-result__primary-subtitle t-14 t-black t-normal")
    for workplace in workplace_results:
        try:
            wp = workplace.text.strip("\n")
            workplaces.append(wp)
            cleaned_wp = re.search("(?<=at).",wp)
            cleaned_workplaces.append(cleaned_wp.group(0))
        except:
            workplaces.append(None)
        try:
            wp = workplace.text.strip("\n")
            wp = wp.strip()
            cleaned_wp = re.search(r"(?<=at ).+",wp)
            cleaned_workplaces.append(cleaned_wp.group(0))
        except:
            cleaned_workplaces.append(None)
```

```python
def scroll_next():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    next_page = driver.find_element(By.CSS_SELECTOR, '[aria-label="Next"]')
    next_page.click()
```

```python
for i in range(5):
    scrape()
    time.sleep(2)
    scroll_next()
    time.sleep(2)
```
